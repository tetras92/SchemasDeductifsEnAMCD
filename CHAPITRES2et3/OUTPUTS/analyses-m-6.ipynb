```python
import pandas as pd
import numpy as np
```

```python
valeurs = [40000, 60000, 80000, 90000, 100000, 120000, 124187]
```

```python
filename = 'recapitulatif-m-6-1-20000.csv'
```

```python
data = pd.read_csv(filename)
```

```python
prec_val = 20001
for val in valeurs:
    tmp_filename = f'recapitulatif-m-6-{prec_val}-{val}.csv'
    prec_val = val + 1
    tmp_data = pd.read_csv(tmp_filename)
    data = pd.concat([data, tmp_data], join='inner')
```

```python
data
```

```python
LesValeurs, CS = np.unique(data["ALL_UNXP"], return_counts=True)
CS = CS.cumsum()
print("les valeurs:", LesValeurs)
print("CS", CS)
PP = np.round(100.*CS/CS[-1], 3)
{LesValeurs[i]: PP[i] for i in range(len(LesValeurs))}
```

```python
LesValeurs, CS = np.unique(data["NB_TO_EXPLAIN"] - data["Base"], return_counts=True)
CS = CS.cumsum()
print("les valeurs:", LesValeurs)
print("CS", CS)
PP = np.round(100.*CS/CS[-1], 3)
{LesValeurs[i]: PP[i] for i in range(len(LesValeurs))}
```

```python
LesValeurs, CS = np.unique(data["NB_TO_EXPLAIN"] - data["Base"] - data["Compo2"], return_counts=True)
CS = CS.cumsum()
print("les valeurs:", LesValeurs)
print("CS", CS)
PP = np.round(100.*CS/CS[-1], 3)
{LesValeurs[i]: PP[i] for i in range(len(LesValeurs))}
```
